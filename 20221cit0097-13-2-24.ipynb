```python
import matplotlib.pyplot as plt
fig = plt.figure()
ax = plt.axes()
import numpy as np
x=np.linspace(0,10,1000)
#print(x)
ax.plot(x,np.sin(x),label='sin',color='red',linestyle='dashed');
ax.plot(x,np.cos(x),label='cos',color='g',linestyle='dotted')
ax.plot(x,np.sin(x+1),label='cos',linestyle='dashdot')
ax.set_title('multiple lines')
ax.set_xlabel('x label')
ax.set_ylabel('y label');
ax.legend()
#ax.legend(loc=1)
plt.show()
```
